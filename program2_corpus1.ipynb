```python
#Importazione dei moduli necessari
import nltk
import math
from collections import Counter
```

```python
#Funzione che legge e restituisce il contenuto del file
def leggiFile(filePath):
    with open(filePath, "r", encoding= "utf-8") as infile:
        contenuto = infile.read()
    return contenuto

#Path del file
file1 = "./Corpus_1.txt"

#Raw text di ciascun file
file1_contenuto = leggiFile(file1)

#Lista di frasi che compongono il testo
frasi1 = nltk.tokenize.sent_tokenize(file1_contenuto)

#Lista di tokens che compongono il testo
tokens1 = nltk.tokenize.word_tokenize(file1_contenuto)

#Lista di PoS Tags sui tokens che compongono il testo
PoSTags1 = nltk.tag.pos_tag(tokens1)

#Lista di bigrammi che compongono il testo
bigrammi1 = nltk.bigrams(tokens1)

#2.1 Estrarre i top 50: Sostantivi, Avverbi e Aggettivi più frequenti (con relativa frequenza ed ordinati in maniera decrescente)
#Funzione che estrae i top 50 Sostantivi, Avverbi e Aggettivi più frequenti, con relativa frequenza ed ordinati in senso decrescente
def Top50_Elementi(lista_TokenPoS, lista_Tag):
    lista_TokenTag = []
    for token, PoS in lista_TokenPoS:
        for Tag in lista_Tag:
            if PoS == Tag:
                lista_TokenTag.append(token)
    frequenze = nltk.FreqDist(lista_TokenTag)
    return frequenze.most_common(50)

#Top 50 Sostantivi, Avverbi e Aggettivi più frequenti, con relativo valore ed ordinati in senso decrescente
top50_Nomi1 = Top50_Elementi(PoSTags1, ["NN", "NNP", "NNPS", "NNS"])
top50_Avverbi1 = Top50_Elementi(PoSTags1, ["RB", "RBR", "RBS"])
top50_Aggettivi1 = Top50_Elementi(PoSTags1, ["JJ", "JJR", "JJS"])

print("2.1: Top 50 Sostantivi, Avverbi e Aggettivi più frequenti")
print(f"Top 50 Sostantivi: {top50_Nomi1}")
print(f"Top 50 Avverbi: {top50_Avverbi1}")
print(f"Top 50 Aggettivi: {top50_Aggettivi1}")
```

```python
#2.2 Estrarre i top 20 n-grammi (per n = [1,2,3,4,5]) più frequenti (con relativa frequenza ed ordinati in maniera decrescente)
#Funzione che estrae i top 20 n-grammi (per n = [1,2,3,4,5]) più frequenti, con relativa frequenza ed ordinati in senso decrescente
def Top20_Ngrammi(tokens, n):
    ngrammi = list(nltk.ngrams(tokens, n))
    frequenze = nltk.FreqDist(ngrammi)
    return frequenze.most_common(20)

#Top 20 n-grammi (per n = [1,2,3,4,5]) più frequenti, con relativo valore ed ordinati in senso decrescente
top20_unigrammi1 = Top20_Ngrammi(tokens1, 1)
top20_bigrammi1 = Top20_Ngrammi(tokens1, 2)
top20_trigrammi1 = Top20_Ngrammi(tokens1, 3)
top20_quadrigrammi1 = Top20_Ngrammi(tokens1, 4)
top20_pentogrammi1 = Top20_Ngrammi(tokens1, 5)

print("2.2: Top 20 n-grammi più frequenti (per n = [1,2,3,4,5])")
print(f"Top 20 unigrammi: {top20_unigrammi1}")
print(f"Top 20 bigrammi: {top20_bigrammi1}")
print(f"Top 20 trigrammi: {top20_trigrammi1}")
print(f"Top 20 quadrigrammi: {top20_quadrigrammi1}")
print(f"Top 20 pentogrammi: {top20_pentogrammi1}")
```

```python
#2.3 Estrarre i top 20 n-grammi (per n = [1,2,3]) di PoS più frequenti (con relativa frequenza ed ordinati in maniera decrescente)
#Funzione che estrae i top 20 n-grammi (per n = [1,2,3]) di PoS più frequenti, con relativa frequenza ed ordinati in senso decrescente
def Top20_Ngrammi_PoS(lista_TokenPoS, n):
    ngrammi = list(nltk.ngrams(lista_TokenPoS, n))
    frequenze = nltk.FreqDist(ngrammi)
    return frequenze.most_common(20)
    
#Top 20 n-grammi (per n = [1,2,3]) di PoS più frequenti, con relativo valore ed ordinati in senso decrescente
top20_unigrammiPoS1 = Top20_Ngrammi_PoS(PoSTags1, 1)
top20_bigrammiPoS1 = Top20_Ngrammi_PoS(PoSTags1, 2)
top20_trigrammiPoS1 = Top20_Ngrammi_PoS(PoSTags1, 3)

print("2.3: Top 20 n-grammi di PoS più frequenti (per n = [1,2,3])")
print(f"Top 20 unigrammi di PoS: {top20_unigrammiPoS1}")
print(f"Top 20 bigrammi di PoS: {top20_bigrammiPoS1}")
print(f"Top 20 trigrammi di PoS: {top20_trigrammiPoS1}")
```

```python
#2.4 b)- Estrarre i top 10 bigrammi di tipo Aggettivo-Sostantivo ed ordinati per probabilità condizionata massima, con relativo valore
#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo
def Top10_Bigrammi_Aggettivo_Nome(tokens):
    bigrammi = list(nltk.bigrams(tokens))
    lista_Bigrammi = [bigramma for bigramma in bigrammi if bigramma[0][1] == ("JJ") and (bigramma[1][1] in {"NN", "NNS", "NNP", "NNPS"})]
    return lista_Bigrammi

#Lista bigrammi estratti
lista_Bigrammi_1 = Top10_Bigrammi_Aggettivo_Nome(PoSTags1)

#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo in base alla probabilità condizionata massima, con relativo valore
def Top10_MaxProbabilitàCondizionata_Bigrammi_Aggettivo_Nome(tokens, lista_Bigrammi):
    vocabolario_Bigrammi = list(set(lista_Bigrammi))
    lista_Bigrammi_ProbabilitàCondizionata = []
    for bigramma in vocabolario_Bigrammi:
        frequenza_Bigramma = tokens.count(bigramma)
        frequenza_PrimoElemento = tokens.count(bigramma[0][0])
        if frequenza_PrimoElemento > 0:
            probabilitàcondizionata = frequenza_Bigramma / frequenza_PrimoElemento
        else:
            probabilitàcondizionata = 0.0
        lista_Bigrammi_ProbabilitàCondizionata.append((bigramma, probabilitàcondizionata))
    lista_Bigrammi_ProbabilitàCondizionata.sort(key=lambda x: x[1], reverse=True)
    return lista_Bigrammi_ProbabilitàCondizionata[:10]

#Top 10 bigrammi di tipo Aggettivo-Sostantivo in base alla probabilità condizionata massima, con relativo valore
top10_maxprobabilitàcondizionata1_bigrammi_aggettivo_nome = Top10_MaxProbabilitàCondizionata_Bigrammi_Aggettivo_Nome(tokens1, lista_Bigrammi_1)

print("2.4 b)- Top 10 bigrammi composti da Aggettivo-Sostantivo e ordinati per probabilità condizionata massima, con relativo valore")
top10_maxprobabilitàcondizionata1_bigrammi_aggettivo_nome
```

```python
#2.4 c)- Estrarre i top 10 bigrammi di tipo Aggettivo-Sostantivo ed ordinati per probabilità congiunta massima, con relativo valore
#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo
def Top10_Bigrammi_Aggettivo_Nome(lista_TokenPoS):
    bigrammi = list(nltk.bigrams(lista_TokenPoS))
    lista_Bigrammi = [bigramma for bigramma in bigrammi if bigramma[0][1] == ("JJ") and (bigramma[1][1] in {"NN", "NNS", "NNP", "NNPS"})]
    return lista_Bigrammi

#Lista bigrammi estratti
lista_Bigrammi_1 = Top10_Bigrammi_Aggettivo_Nome(PoSTags1)

#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo in base alla probabilità congiunta massima, con relativo valore
def Top10_MaxProbabilitàCongiunta_Bigrammi_Aggettivo_Nome(tokens, lista_Bigrammi):
    conteggio_Bigrammi = Counter(lista_Bigrammi)
    lista_Bigrammi_ProbabilitàCongiunta = []
    for bigramma, frequenza_Bigramma in conteggio_Bigrammi.items():
        probabilitàcongiunta = frequenza_Bigramma / len(tokens)
        lista_Bigrammi_ProbabilitàCongiunta.append((bigramma, probabilitàcongiunta))
    lista_Bigrammi_ProbabilitàCongiunta.sort(key=lambda x: x[1], reverse=True)
    return lista_Bigrammi_ProbabilitàCongiunta[:10]

#Top 10 bigrammi di tipo Aggettivo-Sostantivo in base alla probabilità congiunta massima, con relativo valore
top10_maxprobabilitàcongiunta1_bigrammi_aggettivo_nome = Top10_MaxProbabilitàCongiunta_Bigrammi_Aggettivo_Nome(tokens1, lista_Bigrammi_1)

print("2.4 c)- Top 10 bigrammi composti da Aggettivo-Sostantivo e ordinati per probabilità congiunta massima, con relativo valore")
top10_maxprobabilitàcongiunta1_bigrammi_aggettivo_nome
```

```python
#2.4 d)- Estrarre i top 10 bigrammi di tipo Aggettivo-Sostantivo ed ordinati per mutua informazione massima, con relativo valore
#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo
def Top10_Bigrammi_Aggettivo_Nome(lista_TokenPoS):
    bigrammi = list(nltk.bigrams(lista_TokenPoS))
    lista_Bigrammi = [bigramma for bigramma in bigrammi if bigramma[0][1] == ("JJ") and (bigramma[1][1] in {"NN", "NNS", "NNP", "NNPS"})]
    return lista_Bigrammi

#Lista bigrammi estratti
lista_Bigrammi_1 = Top10_Bigrammi_Aggettivo_Nome(PoSTags1)

#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo in base alla mutua informazione massima, con relativo valore
def Top10_MaxMutuaInformazione_Bigrammi_Aggettivo_Nome(tokens, lista_Bigrammi):
    frequenze_SingoliTokens = Counter(tokens)
    conteggio_Bigrammi = Counter(lista_Bigrammi)
    lista_Bigrammi_MI = []
    for bigramma, frequenza_Bigramma in conteggio_Bigrammi.items():
        frequenza_PrimoElemento = frequenze_SingoliTokens[bigramma[0][0]]
        frequenza_SecondoElemento = frequenze_SingoliTokens[bigramma[1][0]]
        if frequenza_PrimoElemento > 0 and frequenza_SecondoElemento > 0:
            MI = math.log(frequenza_Bigramma * len(tokens) / (frequenza_PrimoElemento * frequenza_SecondoElemento), 2)
            lista_Bigrammi_MI.append((bigramma, MI))
    lista_Bigrammi_MI.sort(key=lambda x: x[1], reverse=True)
    return lista_Bigrammi_MI[:10]

#Top 10 bigrammi di tipo Aggettivo-Sostantivo in base alla mutua informazione massima, con relativo valore
top10_maxmutuainformazione1_bigrammi_aggettivo_nome = Top10_MaxMutuaInformazione_Bigrammi_Aggettivo_Nome(tokens1, lista_Bigrammi_1)

print("2.4 d)- Top 10 bigrammi composti da Aggettivo-Sostantivo e ordinati per mutua informazione massima, con relativo valore")
top10_maxmutuainformazione1_bigrammi_aggettivo_nome
```

```python
#2.4 e)- Estrarre i top 10 bigrammi di tipo Aggettivo-Sostantivo ed ordinati per mutua informazione locale massima, con relativo valore
#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo
def Top10_Bigrammi_Aggettivo_Nome(lista_TokenPoS):
    bigrammi = list(nltk.bigrams(lista_TokenPoS))
    lista_Bigrammi = [bigramma for bigramma in bigrammi if bigramma[0][1] == ("JJ") and (bigramma[1][1] in {"NN", "NNS", "NNP", "NNPS"})]
    return lista_Bigrammi

#Lista bigrammi estratti
lista_Bigrammi_1 = Top10_Bigrammi_Aggettivo_Nome(PoSTags1)

#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo in base alla mutua informazione locale massima, con relativo valore
def Top10_MaxMutuaInformazioneLocale_Bigrammi_Aggettivo_Nome(tokens, lista_Bigrammi):
    frequenze_SingoliTokens = Counter(tokens)
    conteggio_Bigrammi = Counter(lista_Bigrammi)
    lista_Bigrammi_LMI = []
    for bigramma, frequenza_Bigramma in conteggio_Bigrammi.items():
        frequenza_PrimoElemento = frequenze_SingoliTokens[bigramma[0][0]]
        frequenza_SecondoElemento = frequenze_SingoliTokens[bigramma[1][0]]
        if frequenza_PrimoElemento > 0 and frequenza_SecondoElemento > 0:
            MI = math.log(frequenza_Bigramma * len(tokens) / (frequenza_PrimoElemento * frequenza_SecondoElemento), 2)
            LMI = frequenza_Bigramma * MI
            lista_Bigrammi_LMI.append((bigramma, LMI))
    lista_Bigrammi_LMI.sort(key=lambda x: x[1], reverse=True)
    return lista_Bigrammi_LMI[:10]

#Top 10 bigrammi di tipo Aggettivo-Sostantivo in base alla mutua informazione locale massima, con relativo valore
top10_maxmutuainformazionelocale1_bigrammi_aggettivo_nome = Top10_MaxMutuaInformazioneLocale_Bigrammi_Aggettivo_Nome(tokens1, lista_Bigrammi_1)

print("2.4 e)- Top 10 bigrammi composti da Aggettivo-Sostantivo e ordinati per mutua informazione locale massima, con relativo valore")
top10_maxmutuainformazionelocale1_bigrammi_aggettivo_nome
```

```python
#2.4 f)- Estrarre i top 10 bigrammi di tipo Aggettivo-Sostantivo ed ordinati in base al numero di elementi comuni per mutua informazione e mutua informazione locale
#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo
def Top10_Bigrammi_Aggettivo_Nome(lista_TokenPoS):
    bigrammi = list(nltk.bigrams(lista_TokenPoS))
    lista_Bigrammi = [bigramma for bigramma in bigrammi if bigramma[0][1] == ("JJ") and (bigramma[1][1] in {"NN", "NNS", "NNP", "NNPS"})]
    return lista_Bigrammi

#Lista bigrammi estratti
lista_Bigrammi_1 = Top10_Bigrammi_Aggettivo_Nome(PoSTags1)

#Funzione che estrae i top 10 bigrammi di tipo Aggettivo-Sostantivo in base al numero di elementi comuni per mutua informazione e mutua informazione locale
def Top10_MaxMutuaInformazione_e_MutuaInformazioneLocale_Bigrammi_Aggettivo_Nome(tokens, lista_Bigrammi):
    frequenze_SingoliTokens = Counter(tokens)
    conteggio_Bigrammi = Counter(lista_Bigrammi)
    lista_Bigrammi_MI = []
    lista_Bigrammi_LMI = []
    for bigramma, frequenza_Bigramma in conteggio_Bigrammi.items():
        frequenza_PrimoElemento = frequenze_SingoliTokens[bigramma[0][0]]
        frequenza_SecondoElemento = frequenze_SingoliTokens[bigramma[1][0]]
        if frequenza_PrimoElemento > 0 and frequenza_SecondoElemento > 0:
            MI = math.log(frequenza_Bigramma * len(tokens) / (frequenza_PrimoElemento * frequenza_SecondoElemento), 2)
            lista_Bigrammi_MI.append((bigramma, MI))
            LMI = frequenza_Bigramma * MI
            lista_Bigrammi_LMI.append((bigramma, LMI))
    lista_Bigrammi_MI.sort(key=lambda x: x[1], reverse=True)
    lista_Bigrammi_LMI.sort(key=lambda x: x[1], reverse=True)
    top10_bigrammi_MI = [bigramma for bigramma, _ in lista_Bigrammi_MI[:10]]
    top10_bigrammi_LMI = [bigramma for bigramma, _ in lista_Bigrammi_LMI[:10]]
    set_LMI = set(top10_bigrammi_LMI)
    elementi_comuni = [b for b in top10_bigrammi_MI if b in set_LMI]
    return elementi_comuni, len(elementi_comuni)

#Top 10 bigrammi di tipo Aggettivo-Sostantivo in base al numero di elementi comuni per mutua informazione e mutua informazione locale
elementi_comuni_1, contatore_elementi_comuni_1 = Top10_MaxMutuaInformazione_e_MutuaInformazioneLocale_Bigrammi_Aggettivo_Nome(tokens1, lista_Bigrammi_1)

print("2.4 f)- Top 10 bigrammi composti da Aggettivo-Sostantivo e ordinati in base al numero di elementi comuni per mutua informazione e mutua informazione locale")
elementi_comuni_1, contatore_elementi_comuni_1
```

```python
#2.5 a)- Considerando le frasi con lunghezza compresa tra 10 e 20 token, in cui almeno la metà (considerare la parte intera della divisione per 2 come valore) dei tokens occorre almeno 2 volte nel corpus. A tal proposito si identifichi la frase con la media della distribuzione di frequenza dei tokens più alta
#Funzione che estrae le frasi con lunghezza compresa tra 10 e 20 token, di cui almeno la metà (considerare la parte intera della divisione per 2 come valore) dei tokens occorre un minimo di 2 volte nel corpus
def LunghezzaFrasi_10_20_Tokens(tokens, frasi):
    frequenze_SingoliTokens = Counter(tokens)
    lista_Frasi_10_20_Tokens = []
    for frase in frasi:
        frasetokenizzata = nltk.tokenize.word_tokenize(frase)
        lunghezzafrase = len(frasetokenizzata)
        if 10 <= lunghezzafrase <= 20:
            soglia_metà = lunghezzafrase // 2
            contatore_noHapax = 0
            for token in frasetokenizzata:
                if frequenze_SingoliTokens[token] >= 2:
                    contatore_noHapax += 1
            if contatore_noHapax >= soglia_metà:
                lista_Frasi_10_20_Tokens.append(frase)
    return lista_Frasi_10_20_Tokens

#Frasi con lunghezza compresa tra 10 e 20 token, di cui almeno la metà (considerare la parte intera della divisione per 2 come valore) dei tokens occorre un minimo di 2 volte nel corpus
lunghezzafrasi_10_20_tokens1 = LunghezzaFrasi_10_20_Tokens(tokens1, frasi1)

#Funzione che estrae la frase con la media della distribuzione di frequenza dei tokens più alta, considerando le frasi con lunghezza compresa tra 10 e 20 token, di cui almeno la metà (considerare la parte intera della divisione per 2 come valore) dei tokens occorre un minimo di 2 volte nel corpus
def Media_MaxDistribuzioneFrequenza_Tokens(tokens, frasi_filtrate):
    frequenze_SingoliTokens = Counter(tokens)
    mediamax_Temporanea = 0
    mediamax_Frase = ""
    for frase in frasi_filtrate:
        frasetokenizzata = nltk.tokenize.word_tokenize(frase)
        lunghezzafrase = len(frasetokenizzata)
        frequenza_totale = sum(frequenze_SingoliTokens[token] for token in frasetokenizzata)
        mediamax = round(frequenza_totale / lunghezzafrase, 3)
        if mediamax > mediamax_Temporanea:
            mediamax_Temporanea = mediamax
            mediamax_Frase = frase
    return mediamax_Frase, mediamax_Temporanea

#Frase con la media della distribuzione di frequenza dei token più bassa
mediamax_Frase_1, mediamax_1 = Media_MaxDistribuzioneFrequenza_Tokens(tokens1, lunghezzafrasi_10_20_tokens1)

print(f"2.5 a)- Frase con la media della distribuzione di frequenza dei token più alta: {Media_MaxDistribuzioneFrequenza_Tokens(tokens1, lunghezzafrasi_10_20_tokens1)}")
```

```python
#2.5 c)- Considerando le frasi con lunghezza compresa tra 10 e 20 token, in cui almeno la metà (considerare la parte intera della divisione per 2 come valore) dei tokens occorre almeno 2 volte nel corpus. A tal proposito si identifichi la frase con la probabilità più alta secondo un modello di Markov del secondo ordine costruito a partire dal corpus in input
#Funzione che estrae le frasi con lunghezza compresa tra 10 e 20 token, di cui almeno la metà (considerare la parte intera della divisione per 2 come valore) dei tokens occorre un minimo di 2 volte nel corpus
def LunghezzaFrasi_10_20_Tokens(tokens, frasi):
    frequenze_SingoliTokens = Counter(tokens)
    lista_Frasi_10_20_Tokens = []
    for frase in frasi:
        frasetokenizzata = nltk.tokenize.word_tokenize(frase)
        lunghezzafrase = len(frasetokenizzata)
        if 10 <= lunghezzafrase <= 20:
            soglia_metà = lunghezzafrase // 2
            contatore_noHapax = 0
            for token in frasetokenizzata:
                if frequenze_SingoliTokens[token] >= 2:
                    contatore_noHapax += 1
            if contatore_noHapax >= soglia_metà:
                lista_Frasi_10_20_Tokens.append(frase)
    return lista_Frasi_10_20_Tokens

#Frasi con lunghezza compresa tra 10 e 20 token, di cui almeno la metà (considerare la parte intera della divisione per 2 come valore) dei tokens occorre un minimo di 2 volte nel corpus
lunghezzafrasi_10_20_tokens1 = LunghezzaFrasi_10_20_Tokens(tokens1, frasi1)

#Funzioni che estraggono la frase con la probabilità più alta secondo un modello di Markov dell'ordine 2 costruito a partire da corpus in input, considerando le frasi con lunghezza compresa tra 10 e 20 token, di cui almeno la metà (considerare la parte intera della divisione per 2 come valore) dei tokens occorre un minimo di 2 volte nel corpus
def ProbabilitàCondizionata_Markov2(trigramma, conteggio_trigrammi, conteggio_bigrammi):
    contesto_bigramma = (trigramma[0], trigramma[1])
    frequenza_trigramma = conteggio_trigrammi[trigramma]
    frequenza_contesto = conteggio_bigrammi[contesto_bigramma]
    if frequenza_contesto > 0:
        return frequenza_trigramma / frequenza_contesto
    else:
        return 0.000001

def MaxProbabilità_Frase_Markov2(tokens, frasi_filtrate):
    conteggio_unigrammi = Counter(tokens)
    conteggio_bigrammi = Counter(nltk.bigrams(tokens))
    conteggio_trigrammi = Counter(nltk.trigrams(tokens))
    probabilitàmax_Temporanea = -1.0
    probabilitàmax_Frase = ""
    for frase in frasi_filtrate:
        frasetokenizzata = nltk.tokenize.word_tokenize(frase)
        if len(frasetokenizzata) < 3:
            continue
        probabilità_PrimaParola = conteggio_unigrammi[frasetokenizzata[0]] / len(tokens)
        contesto_SecondaParola = conteggio_unigrammi[frasetokenizzata[0]]
        probabilità_SecondaParola = (conteggio_bigrammi[(frasetokenizzata[0], frasetokenizzata[1])] / contesto_SecondaParola) if contesto_SecondaParola > 0 else 0.000001
        probabilitàfrase_Temporanea = probabilità_PrimaParola * probabilità_SecondaParola
        frase_trigrammi = list(nltk.trigrams(frasetokenizzata))
        for trigramma in frase_trigrammi:
            probabilitàcondizionata = ProbabilitàCondizionata_Markov2(trigramma, conteggio_trigrammi, conteggio_bigrammi)
            probabilitàfrase_Temporanea *= probabilitàcondizionata
            if probabilitàfrase_Temporanea > probabilitàmax_Temporanea:
                probabilitàmax_Temporanea = probabilitàfrase_Temporanea
                probabilitàmax_Frase = frase
    return probabilitàmax_Frase, probabilitàmax_Temporanea

#Frase con la probabilità più alta secondo un modello di Markov dell'ordine 2
probabilitàmax_Frase_1, probabilitàmax_1 = MaxProbabilità_Frase_Markov2(tokens1, lunghezzafrasi_10_20_tokens1)

print(f"2.5 c)- Frase con la probabilità più alta secondo un modello di Markov dell'ordine 2 costruito a partire dal corpus in input: {MaxProbabilità_Frase_Markov2(tokens1, lunghezzafrasi_10_20_tokens1)}")
```

```python
#2.6 Estratte le Name Entities del testo, identificare per ciascuna classe i 15 elementi più frequenti, ordinati per frequenza decrescente e con relativa frequenza
#Funzione che estrae le Entità Nominate del testo, identificando per ciascuna classe i 15 elementi più frequenti, ordinati per frequenza decrescente e con relativo valore
def Top15_EntitàNominate(lista_TokenPoS, classi_di_interesse):
    albero_NE = nltk.ne_chunk(lista_TokenPoS)
    conteggi_classi = {classe: Counter() for classe in classi_di_interesse}
    for nodo in albero_NE:
        if hasattr(nodo, "label"):
            tipo_NE = nodo.label()
            if tipo_NE in conteggi_classi:
                token_NE = " ".join([token for token, _ in nodo.leaves()])
                conteggi_classi[tipo_NE][token_NE] += 1
    risultati_top15 = {}
    for classe, counter_specifico in conteggi_classi.items():
        risultati_top15[classe] = counter_specifico.most_common(15)
    return risultati_top15

#Definizione delle classi relative alle Name Entities
classi_target = ["GPE", "ORGANIZATION", "PERSON"]

#Estrazione totale delle Name Entities
dizionario_top15 = Top15_EntitàNominate(PoSTags1, classi_target)

#Entità Nominate del testo, identificando per ciascuna classe i 15 elementi più frequenti, ordinati per frequenza decrescente e con relativo valore
top15_GPE1 = dizionario_top15["GPE"]
top15_ORGANIZATION1 = dizionario_top15["ORGANIZATION"]
top15_PERSON1 = dizionario_top15["PERSON"]

print(f"2.6: Entità Nominate del testo, con i 15 elementi più frequenti per ciascuna classe")
print(f"Top 15 elementi della classe 'GPE': {top15_GPE1}")
print(f"Top 15 elementi della classe 'ORGANIZATION': {top15_ORGANIZATION1}")
print(f"Top 15 elementi della classe 'PERSON': {top15_PERSON1}")
```
