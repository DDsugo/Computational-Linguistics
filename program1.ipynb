```python
#Importazione dei moduli necessari
import nltk
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
from string import punctuation
```

```python
#Funzione che legge e restituisce il contenuto del file
def leggiFile(filePath):
    with open(filePath, "r", encoding= "utf-8") as file:
        contenuto = file.read()
    return contenuto

#Path dei file
file1 = "./Corpus_1.txt"
file2 = "./Corpus_2.txt"

#Raw text di ciascun file
file1_contenuto = leggiFile(file1)
file2_contenuto = leggiFile(file2)

#Lista di frasi che compongono il testo
frasi1 = nltk.tokenize.sent_tokenize(file1_contenuto)
frasi2 = nltk.tokenize.sent_tokenize(file2_contenuto)

#Lista di tokens che compongono il testo
tokens1 = nltk.tokenize.word_tokenize(file1_contenuto)
tokens2 = nltk.tokenize.word_tokenize(file2_contenuto)

#Numero di frasi che compongono il testo
numerofrasi1 = len(frasi1)
numerofrasi2 = len(frasi2)

#Numero di tokens che compongono il testo
numerotokens1 = len(tokens1)
numerotokens2 = len(tokens2)

#1.1 Confrontare i due corpus rispetto a Numero di frasi e Numero di tokens
print("1.1: Numero di frasi e token")
print(f"Il primo corpus è composto da {numerofrasi1} frasi")
print(f"Il secondo corpus è composto da {numerofrasi2} frasi")
print(f"Il primo corpus è composto da {numerotokens1} tokens")
print(f"Il secondo corpus è composto da {numerotokens2} tokens")
```

```python
#1.2: Lunghezza media delle frasi in token e lunghezza media dei token in caratteri, escludendo la punteggiatura
#Lunghezza media delle frasi in token
lunghezzamediafrasi1_inTokens = numerotokens1/numerofrasi1
lunghezzamediafrasi2_inTokens = numerotokens2/numerofrasi2

#Funzione che calcola la lunghezza media dei token in caratteri, escludendo la punteggiatura
def lunghezzaMediaTokens_inCaratteri_noPunteggiatura(corpus):
    tokens = nltk.tokenize.word_tokenize(corpus)
    tokens_noPunteggiatura = [token for token in tokens if token not in punctuation]
    lunghezzamedia_tokens = sum([len(token) for token in tokens_noPunteggiatura]) / len(tokens_noPunteggiatura)
    return lunghezzamedia_tokens

#Lunghezza media dei tokens in caratteri, escludendo la punteggiatura
lunghezzamediatokens1_inCaratteri_noPunteggiatura = lunghezzaMediaTokens_inCaratteri_noPunteggiatura(file1_contenuto)
lunghezzamediatokens2_inCaratteri_noPunteggiatura = lunghezzaMediaTokens_inCaratteri_noPunteggiatura(file2_contenuto)

print("1.2: Lunghezza media delle frasi in token e lunghezza media dei token in caratteri, escludendo la punteggiatura")
print(f"Il primo corpus ha una lunghezza media di frase di {lunghezzamediafrasi1_inTokens:.2f} tokens")
print(f"Il secondo corpus ha una lunghezza media di frase di {lunghezzamediafrasi2_inTokens:.2f} tokens")
print(f"Il primo corpus ha una lunghezza media di token di {lunghezzamediatokens1_inCaratteri_noPunteggiatura:.2f} caratteri")
print(f"Il secondo corpus ha una lunghezza media di token di {lunghezzamediatokens2_inCaratteri_noPunteggiatura:.2f} caratteri")
```

```python
#1.3: Numero di hapax tra i primi 500, 1000, 3000 token, e nell'intero corpus
#Funzione che calcola il numero di hapax tra i primi 500, 1000, 3000 tokens, e l'intero corpus
def numeroHapax_perPorzione(tokens):
    lista_porzioniIncrementali = [500, 1000, 3000, len(tokens)]
    for porzione in lista_porzioniIncrementali:
        hapax_counter = sum(1 for token in tokens[:porzione] if tokens.count(token) == 1)
        print(f"Numero di hapax nei primi {porzione} token: {hapax_counter}")

print("1.3: Numero di hapax tra i primi 500, 1000, 3000 token, e nell'intero corpus")
print("Corpus 1:")
numerohapax1_perPorzione = numeroHapax_perPorzione(tokens1)
print()
print("Corpus 2:")
numerohapax2_perPorzione = numeroHapax_perPorzione(tokens2)
```

```python
#1.4 Confrontare i due corpus rispetto a Dimensione del vocabolario e Ricchezza lessicale, entrambe calcolate per porzioni incrementali di 200 tokens
#Funzione che calcola la dimensione del vocabolario e la ricchezza lessicale per porzioni incrementali di 200 tokens fino alla totalità del corpus
def dimensioneVocabolario_e_TTR_incrementale200tokens(tokens):
    incremento = 200
    while incremento < len(tokens):
        porzione = tokens[0:incremento]
        vocabolario_porzione = set(porzione)
        TTR_porzione = len(vocabolario_porzione) / len(porzione)
        print(f"Dimensione del vocabolario per i primi {incremento} tokens del corpus: {len(vocabolario_porzione)}")
        print(f"Indice di ricchezza lessicale per i primi {incremento} tokens del corpus: {TTR_porzione:.2f}")
        incremento += 200

print("1.4: Dimensione del vocabolario e ricchezza lessicale, calcolata per porzioni incrementali di 200 token fino ad arrivare a tutto il testo")
print("Corpus 1:")
dimensioneVocabolario_e_TTR1_incrementale200tokens = dimensioneVocabolario_e_TTR_incrementale200tokens(tokens1)
print()
print("Corpus 2:")
dimensioneVocabolario_e_TTR2_incrementale200tokens = dimensioneVocabolario_e_TTR_incrementale200tokens(tokens2)
```

```python
#1.5 Confrontare i due corpus rispetto a Numero di lemmi distinti
#Funzione che calcola il numero di lemmi distinti
def numeroLemmi_distinti(tokens):
    lemmatizzatore = WordNetLemmatizer()
    lemmidistinti = set([lemmatizzatore.lemmatize(token) for token in tokens])
    return len(lemmidistinti)

#Numero di lemmi distinti
numerolemmi1 = numeroLemmi_distinti(tokens1)
numerolemmi2 = numeroLemmi_distinti(tokens2)

print("1.5: Numero di lemmi distinti")
print(f"Il primo corpus è composto da {numerolemmi1} lemmi distinti")
print(f"Il secondo corpus è composto da {numerolemmi2} lemmi distinti")
```

```python
#1.6 Confrontare i due corpus rispetto a distribuzione di frasi con polarità positiva e negativa
#Funzione che calcola la distribuzione di frasi con polarità positiva e negativa
def distribuzionePolarità(frasi):
    polarità_positiva = 0
    polarità_negativa = 0
    for frase in frasi:
        analisi = TextBlob(frase)
        if analisi.sentiment.polarity > 0:
            polarità_positiva += 1
        elif analisi.sentiment.polarity < 0:
            polarità_negativa += 1
    return polarità_positiva, polarità_negativa

#Distribuzione di frasi con polarità positiva e negativa
distribuzionepolarità1 = distribuzionePolarità(frasi1)
distribuzionepolarità2 = distribuzionePolarità(frasi2)

print("1.6: Distribuzione di frasi con polarità positiva e negativa")
print(f"Il primo corpus è composto da {distribuzionepolarità1[0]} frasi con polarità positiva e {distribuzionepolarità1[1]} frasi con polarità negativa")
print(f"Il secondo corpus è composto da {distribuzionepolarità2[0]} frasi con polarità positiva e {distribuzionepolarità2[1]} frasi con polarità negativa")
```
